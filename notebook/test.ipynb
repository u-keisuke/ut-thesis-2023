```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../log.txt', header=None, sep=',')
df.columns = ["seconds", "iteration", "exploitability", "br0-ev1", "ev0-br1"]
```

```python
# plot dots without lines
plt.plot(df['iteration'], df['br0-ev1'], 'o', color='blue', )
plt.plot(df['iteration'], df['ev0-br1'], 'o', color='orange')
plt.axhline(y=-1/18)
plt.legend(['br0-ev1', 'ev0-br1'])
plt.xlabel('iteration')
plt.ylabel('utility for player 0')
plt.xscale('log')
```
